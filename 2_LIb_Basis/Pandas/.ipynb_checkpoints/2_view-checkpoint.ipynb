```python
import pandas as pd
```

```python
col_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT','MEDV']
```

```python
#read_csv arg1:文件路径及名称.与代码同文件夹可以省略路径 arg2:列名称 arg3:列间隔  arg4:添加的列信息
data = pd.read_csv('data.csv', header = None, sep = r'\s+', names = col_names)
```

```python
import matplotlib.pyplot as plt
```

```python
data.INDUS.plot(kind = 'line', color = 'r', label = 'INDUS', linewidth = 1, grid = True, linestyle = '-')
data.AGE.plot(kind = 'line', color = 'g', label = 'AGE', linewidth = 1, grid = True, linestyle = '-')

plt.legend()
plt.title('Line plot of INDUE & AGE')
plt.xlabel('x: Range Index')
plt.ylabel('y: INDUS & AGE Value')

plt.show()
```

```python
#绘制直方图
```

```python
#arg1：hist表示直方图 arg2:bins
data.RM.plot(kind = 'hist', bins = 50)
plt.legend()
plt.title('Histogram of RM')
plt.xlabel('x: RM')
plt.ylabel('y: count')

plt.show()
```

```python
#绘制累计直方图
```

```python
data.RM.plot(kind = 'hist', bins = 50, cumulative = True)
plt.legend()
plt.title('Cumulative histogram of RM')
plt.xlabel('x: RM')
plt.ylabel('y: Cummulative count')

plt.show()
```

```python
#散点图
```

```python
data.plot(kind = 'scatter', x = 'TAX', y = 'PTRATIO', color = 'r')

plt.title('Scatter histogram of RM')
plt.xlabel('x: TAX')
plt.ylabel('y: PTRATIO')

plt.show()
```
