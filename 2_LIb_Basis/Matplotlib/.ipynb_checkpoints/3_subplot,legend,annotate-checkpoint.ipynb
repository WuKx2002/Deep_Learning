```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
```

```python
y1 = x + 0.5
y2 = x ** 2
y3 = np.log(2 * x + 1)
y4 = (1 / (x + 1)) * np.sin(x)
```

```python
#arg1:子图行数  arg2:子图列数  arg3:子图id
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'r')
plt.subplot(2, 2, 2)
plt.plot(x, y2, 'y')
plt.subplot(2, 2, 3)
plt.plot(x, y3, 'g')
plt.subplot(2, 2, 4)
plt.plot(x, y4, 'b')
plt.show()
```

```python
#arg1:子图行数  arg2:子图列数  arg3:子图id label:备注信息
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'ro--',label = 'x + 0.5' )
plt.legend() #默认不遮挡图像的位置

plt.subplot(2, 2, 2)
plt.plot(x, y2, 'y-',label = 'x^2')
plt.legend(loc = 'upper right') #备注信息显示在左上角

plt.subplot(2, 2, 3)
plt.plot(x, y3, 'gs',label = 'log(2x+1)')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, y4, 'b--',label = 'sinx/ (x + 1)')
plt.legend()
plt.show()
```

```python
#添加注释
plt.subplot(2, 2, 4)
plt.plot(x, y4, 'bs-',label = 'sinx/ (x + 1)')
plt.legend(loc = 'lower left')
#arg1:显示内容 arg2:箭头指向的位置 arg3:标注文本的位置 arg4:字体大小 arg5:箭头属性
plt.annotate('max_value', xy = (1, 0.42), xytext = (2, 0.4), fontsize = 10, arrowprops = dict(facecolor = 'red', width = 5)) 
plt.show()
```
