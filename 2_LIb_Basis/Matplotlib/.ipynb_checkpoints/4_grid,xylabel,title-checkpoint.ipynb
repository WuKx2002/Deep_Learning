```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(20)
y = np.sin(x)
```

```python
plt.plot(x, y, 'bo--', label = 'sinx')
plt.legend()
plt.xticks(np.arange(20))
plt.grid(True) #grid方法用于设置是否开启辅助网格线
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('The sin(x) figure')

plt.show()
```
