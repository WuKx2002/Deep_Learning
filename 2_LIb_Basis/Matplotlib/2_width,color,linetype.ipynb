```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
z = range(10)
print(x)
print(z)
```

```python
y1 = x + 2000
y2 = x ** 4
y3 = np.exp(x)
```

```python
plt.plot(x, y1, linewidth = 2 )
plt.plot(x, y2, linewidth = 4)
plt.plot(x, y3, linewidth = 6)

plt.show()
```

```python
plt.plot(x, y1, 'r', linewidth = 2 )
plt.plot(x, y2, 'y', linewidth = 4)
plt.plot(x, y3, 'b', linewidth = 6)

plt.show()
```

```python
plt.plot(x, y1, 'o-')
plt.plot(x, y2, 'yo--')
plt.plot(x, y3, 'b^-')
plt.show()
```

```python
plt.plot(x, y1, '-.')
plt.plot(x, y2, '--')
plt.plot(x, y3, 's')
plt.show()
```
